```python
import pandas as pd
import numpy as np
import math
import re
```

```python
df = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', engine='python')
df.head()
```

```python
# Get column names
column_names = df.columns
print(column_names)
# Get column data types
print(df.dtypes)
# Also check if the column is unique
for i in column_names:
  print('{} is unique: {}'.format(i, df[i].is_unique))
```

```python
df.rename(columns={'COST (£) charged to Wellcome (inc VAT when charged)':'Cost'},inplace=True)
df.head()
```

```python
pattern = re.compile(r'£')

for match in df['Cost']:
    matches = re.findall(pattern, match)
    print(matches)
```

```python
df['Cost'] = df['Cost'].apply(lambda x: x.strip('£'))
df['Cost'] = df['Cost'].apply(lambda x: x.strip('$'))
df['Cost'] = df['Cost'].apply(lambda x: float(x))
```

```python
df['Journal title'].unique()
```

```python
pattern = re.compile(r'([Tt][Hh][Ee])?(\s)?([J][a-z]*)\s([of]{2})?(\s)?([M][a-z]+)\s(C\w+).?')

for match in df['Journal title']:
    match = str(match)
    matches = re.findall(pattern, match)
    print(matches)
```

```python
matches = []
for match in df['Journal title']:
    match = str(match)
    matches.append(re.sub(pattern, r'Journal of Medicinal Chemistry', match))
df['Journal title'] = matches
```

```python
pattern = re.compile(r'(([J][a-z]*)\s([of]{2})?\s?([O][a-z]+)\s(C\w+))')

matches = []
for match in df['Journal title']:
    match = str(match)
    matches.append(re.sub(pattern, r'Journal of Organic Chemistry', match))
df['Journal title'] = matches
df['Journal title'].value_counts()
```

```python
pattern = re.compile(r'([Pp][Ll][Oo][Ss])(.*)?')
```

```python
matches = []
for match in df['Journal title']:
    match = str(match)
    matches.append(re.sub(pattern, r'Public Library of Science', match))
df['Journal title'] = matches
df['Journal title'].value_counts()
```

```python
pattern = re.compile(r'&')

matches=[]
for match in df['Journal title']:
    match = str(match)
    matches.append(re.sub(pattern, r'and', match))
df['Journal title'] = matches
df['Journal title'].value_counts()
```

```python
pattern = re.compile(r'(the)?(j.*)\s?(of)?\s?(biol\w*)?\s?(chem\w*).?', re.IGNORECASE)

matches = []
for match in df['Journal title']:
    match = str(match)
    matches.append(re.sub(pattern, r'Journal of Biological Chemistry', match))
df['Journal title'] = matches
df['Journal title'].value_counts()
```

```python
pattern = re.compile(r'(.*)pnas(.*)', re.IGNORECASE)

matches = []
for match in df['Journal title']:
    match = str(match)
    matches.append(re.sub(pattern, r'Proceedings of the National Academy of Sciences', match))
df['Journal title'] = matches
df['Journal title'].value_counts()
```

```python
pattern = re.compile(r'(nucleic)\s(acids?)\s(research)', re.IGNORECASE)

matches = []
for match in df['Journal title']:
    match = str(match)
    matches.append(re.sub(pattern, r'Nucleic Acids Research', match))
df['Journal title'] = matches
df['Journal title'].value_counts()
```

```python
pattern = re.compile(r'(hum\w*)\s(mol.*)', re.IGNORECASE)

matches = []
for match in df['Journal title']:
    match = str(match)
    matches.append(re.sub(pattern, r'Human Molecular Genetics', match))
df['Journal title'] = matches
df['Journal title'].value_counts()
```
