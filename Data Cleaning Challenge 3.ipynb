# 1. Consider the two variables: TOTAL_REVENUE and TOTAL_EXPENDITURE. Do these variables have outlier values?

```python
plt.figure(figsize = (15,5))

plt.subplot(1,2,1)
plt.boxplot(education_df["TOTAL_REVENUE"], whis = 5)
plt.title("Box plot of total revenue")

plt.subplot(1,2,2)
plt.boxplot(education_df["TOTAL_EXPENDITURE"], whis = 5)
plt.title("Box plot of total expenditure")
plt.show()
```

Both boxplots have outliers in the higher ranges.

# 2. If you detect outliers in the TOTAL_REVENUE and TOTAL_EXPENDITURE variables, apply the techniques you learned in this checkpoint to eliminate them and validate that there's no outlier values afterwards you handled them.

Because of the nature of these values, we will be limiting the higher end of the graph and bringing those values down. The method we will be using is the winsorization method, one-way.

```python
from scipy.stats.mstats import winsorize

winsorized_revenue = winsorize(education_df["TOTAL_REVENUE"], (0, 0.03))

winsorized_expenditure = winsorize(education_df['TOTAL_EXPENDITURE'], (0, 0.03))
```

```python
plt.figure(figsize = (15,5))

plt.subplot(1,2,1)
plt.boxplot(winsorized_revenue, whis = 5)
plt.title("Box plot of winsorized total revenue")

plt.subplot(1,2,2)
plt.boxplot(winsorized_expenditure, whis = 5)
plt.title("Box plot of winsorized total expenditure")
plt.show()
```

If we limit the values to 97% of the current dataframe, we no longer have any outliers

# 3. Create another variable by subtracting the original TOTAL_EXPENDITURE from TOTAL_REVENUE (before you eliminated the outliers). You can think of it as a kind of budget deficit in education. Do you find any outlier values in this new variable? If so, eliminate them using the technique you think most suitable.

```python
budget_deficit = education_df.TOTAL_REVENUE - education_df.TOTAL_EXPENDITURE
```

```python
plt.boxplot(budget_deficit, whis = 5)
plt.title("Box plot of budget deficit")
```

We have outliers on both sides of the plot. Can use winsorizations once again

```python
winsorized_budget = winsorize(budget_deficit, (0.02, 0.02))

plt.boxplot(winsorized_budget, whis = 5)
plt.title("Box plot of budget deficit")
```

Now there are no longer any outliers on either side after only keeping 96% of the values.
